```python
import os

tr_path = 'data/affectnet/train_class/'
cl_dirs = sorted(os.listdir(tr_path))
files = []
for _dir in cl_dirs:
    cl_files = sorted(os.listdir(tr_path + _dir))
    files += [tr_path + _dir + '/' + file for file in cl_files]
    
import numpy as np
files = np.array(files)
```

```python
import numpy as np
import pickle as pkl

idx = np.random.permutation(len(files))
p_size = 3000

for i in range(len(files) // p_size):
    p_idx = idx[i * p_size:(i + 1) * p_size]
    
    p_files = files[p_idx]
    p_ims = []
    for file in p_files:
        im = cv2.imread(file)
        p_ims.append(im)
    
    with open('data/affectnet/train/' + str(i), 'wb+') as f:
        pkl.dump(p_ims)
    print(i)
    
p_idx = idx[- (len(files) % p_size):]

p_files = files[p_idx]
p_ims = []
for file in p_files:
    im = cv2.imread(file)
    p_ims.append(im)
with open('data/affectnet/train/' + str(i), 'wb+') as f:
    pkl.dump(p_ims)

print(i)
```
