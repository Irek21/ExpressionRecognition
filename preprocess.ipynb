```python
import os

tr_path = 'data/affectnet/train_class/'
cl_dirs = sorted(os.listdir(tr_path))
files = []
labels = []
for i, _dir in enumerate(cl_dirs):
    cl_files = sorted(os.listdir(tr_path + _dir))
    files += [tr_path + _dir + '/' + file for file in cl_files]
    labels += [i for j in range(len(cl_files))]
    
import numpy as np
files = np.array(files)
labels = np.array(labels)
```

```python
files
```

```python
import numpy as np
import pickle as pkl

idx = np.random.permutation(len(files))
# idx = np.arange(len(files))
p_size = 1000

for i in range(20): # (len(files) // p_size):
    p_idx = idx[i * p_size:(i + 1) * p_size]
    
    p_files = files[p_idx]
    p_labels = labels[p_idx]
    p_ims = []
    for j, file in enumerate(p_files):
        im = cv2.imread(file)
        p_ims.append((im, p_labels[j]))
    
    with open('data/affectnet/train/' + str(i), 'wb+') as f:
        pkl.dump(p_ims, f)
    print(i)
```

```python
from itertools import groupby 
import cv2

class PackLoader():
    def __init__(self, psize, path):
        
        self.path = path
        files = [path + '/images/'+ imfile for imfile in os.listdir(path + '/images')]
        lfiles = [path + '/annotations/' + file[:-4] + '_exp.npy' for file in os.listdir(path + '/images')]
            
        self.files = np.array(files)
        self.labels = np.array([int(np.load(labfile).item()) for labfile in lfiles])
        self.idx = np.random.permutation(len(files))
        self.psize = psize

        return

    def pack(self, i):
        idx = self.idx[i * self.psize:(i + 1) * self.psize]
    
        files = self.files[idx]
        labels = self.labels[idx]
        pack = []
        for j, file in enumerate(files):
            im = cv2.imread(file)
            pack.append((im, labels[j]))
            
        return pack
```

```python
root = 'data/fer2013/'
import numpy as np
```

```python
import shutil as shtl
import os

cl_dirs = sorted(os.listdir(root + 'train'))
os.mkdir(root + 'train/images')

offs = 0
labels = np.array([])
for cl, dir_ in enumerate(cl_dirs):
    files = os.listdir(root + 'train/' + dir_)
    
    for i, file in enumerate(files):
        os.rename(root + 'train/' + dir_ + '/' + file, 
                  root + 'train/' + dir_ + '/' + str(offs + i))
        shtl.move(root + 'train/' + dir_ + '/' + str(offs + i), root + 'train/images')

    offs += len(files)
    shtl.rmtree(root + 'train/' + dir_)
    
    cl_labels = np.zeros(len(files)) + cl
    labels = np.concatenate([labels, cl_labels])
```
